```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import data
from skimage.filters import try_all_threshold
%config InlineBackend.figure_format = 'retina'


img = data.camera()
#plt.imshow(img)

fig,ax = try_all_threshold(img,figsize=(6,8),verbose=False)
plt.show()
```

```python
## Now let's show an example of how otsu's method works 
```

```python
from skimage.filters import threshold_otsu


image= data.camera()
thresh = threshold_otsu(image)
binary = image > thresh

fig, axis = plt.subplots(ncols = 3, figsize=(10,2.5))
ax = axis.ravel()

ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3)

ax[0].imshow(image,cmap=plt.cm.gray)
ax[0].axis('off')                         

ax[1].hist(image.ravel(),bins=256,color = "skyblue")
ax[1].axvline(thresh,color='darkorange')                        

ax[2].imshow(binary,cmap=plt.cm.gray)
ax[2].axis('off') 
```

```python
# Try the same operation with different image

from skimage.color import rgb2gray
from skimage.transform import rescale, resize

im = data.chelsea()
image_cat= rgb2gray(im)
image_cat = resize(image_cat,image.shape)
thresh_cat = threshold_otsu(image_cat)
binary_cat = image_cat > thresh_cat


fig, axis = plt.subplots(nrows=2,ncols = 3, figsize=(10,4))
ax = axis.ravel()

ax[0] = plt.subplot(2, 3, 1)
ax[1] = plt.subplot(2, 3, 2)
ax[2] = plt.subplot(2, 3, 3)

ax[3] = plt.subplot(2, 3, 4)
ax[4] = plt.subplot(2, 3, 5)
ax[5] = plt.subplot(2, 3, 6)


ax[0].imshow(image,cmap=plt.cm.gray)
ax[0].axis('off')                         

ax[1].hist(image.ravel(),bins=256,color = "skyblue")
ax[1].axvline(thresh,color='darkorange')                        

ax[2].imshow(binary,cmap=plt.cm.gray)
ax[2].axis('off')

ax[3].imshow(image_cat,cmap=plt.cm.gray)
ax[3].axis('off')                         

ax[4].hist(image_cat.ravel(),bins=256,color = "skyblue")
ax[4].axvline(thresh_cat,color='darkorange')                        

ax[5].imshow(binary_cat,cmap=plt.cm.gray)
ax[5].axis('off') 

print(axis.shape)

```
